## Create a subset from VRSBench based on question types

```python
import json
from collections import defaultdict

# Path to your full VQA data
json_path = "/home/vol-llm/datasets/ISR/VRSBench/validation/VRSBench_EVAL_vqa.json"

# Output path
output_path = "label.json"

# Question types to extract
target_types = {
    "object color",
    "object existence",
    "object category",
    "object shape",
    "scene type",
    "object quantity",
    "object position",
    "object direction",
    "object size",
    "reasoning"
}

# Load the full dataset
with open(json_path, "r", encoding="utf-8") as f:
    vqa_data = json.load(f)

# Prepare storage for each type
type_samples = defaultdict(list)

# Specify the number of samples to collect per type
# TODO: change if needed
NUM_SAMPLES_PER_TYPE = 100

# Iterate and collect up to 100 per type
for item in vqa_data:
    q_type = item["type"]
    if q_type in target_types and len(type_samples[q_type]) < NUM_SAMPLES_PER_TYPE:
        type_samples[q_type].append(item)

    # Stop if all types are filled
    if all(len(type_samples[t]) >= NUM_SAMPLES_PER_TYPE for t in target_types):
        break

# Flatten into a single list for saving
final_subset = []
for items in type_samples.values():
    final_subset.extend(items)

# Save to JSON
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(final_subset, f, ensure_ascii=False, indent=4)

print(f"Saved {len(final_subset)} samples to {output_path}")
```

## Get the lists of question_id in different question types

```python
import json

# Path to your full VQA data
json_path = "label.json"

# Read json
with open(json_path, "r", encoding="utf-8") as f:
    vqa_data = json.load(f)

# Empty lists
qids_obj_qnty = []
qids_obj_pos = []
qids_obj_dir = []
qids_obj_size = []
qids_reasoning = []
qids_obj_color = []
qids_obj_exist = []
qids_obj_cat = []
qids_obj_shape = []
qids_scene_type = []

# Iterate over json
for item in vqa_data:
    if item["type"] == "object quantity":
        qids_obj_qnty.append(item["question_id"])
    elif item["type"] == "object position":
        qids_obj_pos.append(item["question_id"])
    elif item["type"] == "object direction":
        qids_obj_dir.append(item["question_id"])
    elif item["type"] == "object size":
        qids_obj_size.append(item["question_id"])
    elif item["type"] == "reasoning":
        qids_reasoning.append(item["question_id"])
    elif item["type"] == "object color":
        qids_obj_color.append(item["question_id"])
    elif item["type"] == "object existence":
        qids_obj_exist.append(item["question_id"])
    elif item["type"] == "object category":
        qids_obj_cat.append(item["question_id"])
    elif item["type"] == "object shape":
        qids_obj_shape.append(item["question_id"])
    elif item["type"] == "scene type":
        qids_scene_type.append(item["question_id"])

print("=== Question IDs ===")
print(f"Object Quantity: \t{qids_obj_qnty}")
print(f"Object Position: \t{qids_obj_pos}")
print(f"Object Direction: \t{qids_obj_dir}")
print(f"Object Size: \t\t{qids_obj_size}")
print(f"Reasoning: \t\t{qids_reasoning}")
print(f"Object Color: \t\t{qids_obj_color}")
print(f"Object Existence: \t{qids_obj_exist}")
print(f"Object Category: \t{qids_obj_cat}")
print(f"Object Shape: \t\t{qids_obj_shape}")
print(f"Scene Type: \t\t{qids_scene_type}")
```
